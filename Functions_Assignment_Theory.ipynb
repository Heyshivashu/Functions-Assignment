**2. Explain the concept of function arguments and parameters in Python.**

Ans: In Python, functions are a core aspect of the language, allowing for code reusability and modular design. A critical component of functions is the ability to accept data through parameters and arguments. Understanding these concepts is essential for writing effective Python code.

**Parameters** : Parameters are variables that are defined in the function signature. They act as placeholders for the values that will be passed to the function when it is called. Parameters allow functions to accept input and operate on that data. Types of Parameters are:

* Positional Parameters: The order in which arguments are passed to the function must match the order of the parameters in the function definition.
* Default Parameters: Parameters can have default values, which makes them optional. If no argument is provided for a default parameter, the default value is used.
* Keyword Parameters: When calling a function, you can specify the parameter name along with the value, allowing you to pass arguments in any order.
* Variable-Length Parameters: Functions can accept a variable number of positional or keyword arguments using args and kwargs.

```python
def multiply(x, y):  # x and y are parameters
    return x * y
result = multiply(4, 5)
print(result)
```

**Arguments**: Arguments are the actual values that are passed to the function when it is called. They are provided in the parentheses of the function call. Arguments can be literals, variables, or expressions that match the parameters defined in the function.

* Positional Arguments: These are the most common type of arguments, where the order in which they are passed matters.
* Keyword Arguments: In this case, arguments are passed by explicitly naming the parameter, allowing the caller to specify values in any order.
* Default Arguments: Functions can define default values for parameters. If an argument is not provided during the function call, the default value is used. This allows for optional parameters.
* Variable-Length Arguments: Python functions can accept an arbitrary number of arguments using *args for positional arguments and **kwargs for keyword arguments.


```python
def print_book_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")
print_book_info(title="Data Analytics", author="Shivanshu", year=2024)
```

**3. What are the different ways to define and call a function in Python?**

Ans: In Python, we can define and call functions in several ways. Here are the most common methods:

* **Basic Function Definition and Call**: We can define a simple function using the def keyword and call it by its name.

```python
def greet(name):
    return f"Hello, {name}!"
print(greet("Shivanshu"))
```

* **Using Lambda Functions**: For simple, one-line functions, We can use lambda:

```python
square = lambda x: x ** 2
print(square(5))
```

* **Default Parameters**: We can define functions with default parameter values:

```python
def greet(name="Shivanshu"):
    return f"Hello, {name}!"
print(greet())
print(greet("Data Analysts"))
```

* **Variable-Length Arguments** : We can use args for variable-length positional arguments and kwargs for variable-length keyword arguments:

```python
def add(*args):
    return sum(args)

print(add(65, 1, 3))

def display_info(**kwargs):
    return kwargs

print(display_info(name="Shivanshu", age=23))
```

* **Nested Functions**: We can define functions inside other functions:

```python
def outer_function(text):
    def inner_function():
        return f"Inner function says: {text}"
    return inner_function()
print(outer_function("Yo bro"))
```

* **Function Annotations**: We can add type hints to our functions for better readability:

```python
def add(a: int, b: int) -> int:
    return a + b

print(add(67, 2))
```

* **Using the functools Module**: We can use decorators from the functools module, like wraps:

```python
from functools import wraps

def decorator(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print("Before calling the function")
        result = func(*args, **kwargs)
        print("After calling the function")
        return result
    return wrapper

@decorator
def say_hello(name):
    return f"Hello, {name}!"
print(say_hello("Shivanshu Singh Parihar"))
```

**6. Explain the concept of generators in Python and how they are defined.**

Ans: Generators in Python are a special type of iterator that allow us to iterate over a sequence of values without storing the entire sequence in memory. They provide a convenient way to create iterators using a simple syntax. A generator is a function that returns an iterator and allows us to iterate through a sequence of values. Instead of returning a single value, a generator uses the yield statement to produce a series of values over time.
Generators are memory-efficient because they yield items one at a time and only when requested, rather than generating all values at once and storing them in memory.

```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

# Creating a generator
counter = count_up_to(5)

# Using the generator
for number in counter:
    print(number)
```

**How we can define**

We can define a generator function just like a regular function, but instead of using return, we use yield. When the generator function is called, it does not execute the body of the function; instead, it returns a generator object.

```python
squares = (x * x for x in range(5))

# Using the generator expression
for square in squares:
    print(square)
```

**7. What are the advantages of using generators over regular functions?**

Ans: Generators offer several advantages over regular functions, especially when dealing with sequences of data. Here are some advantages are:

* Memory Efficiency: Generators do not store all values in memory at once, making them ideal for working with large datasets or infinite sequences.
* Reduced Memory Footprint: Generators uses significantly less memory compared to storing an entire list or collection of items.

* On Demand Value Generation: Generators can be faster than list comprehensions or regular functions that return lists, as they avoid the overhead of creating and storing large data structures.
* Short-Circuiting: We can stop iteration as soon as we find what we need, without having to compute the entire sequence.
* Cleaner Syntax: The code easier to read and understand compared to managing the state manually with regular functions and returning lists.
* No Need for Helper Variables: Generators encapsulate the iteration logic, reducing the need for additional variables to manage state.
* Infinite Iterators: Generators can be used to create infinite sequences without running out of memory. We can continue generating values as needed without predefining a limit.
* Preserving State: Generators automatically maintain their state between successive calls. This means we do not have to manage the state manually, which simplifies the implementation of complex iteration patterns.
* Pipeline Processing: Generators can be composed together to create data processing pipelines. This allows us to chain multiple generators to transform or filter data without loading everything into memory at once.

```python
#Regular Function
def get_squares(n):
    return [x * x for x in range(n)]

squares = get_squares(1000000)
```

```python
#Generator
def get_squares(n):
    for x in range(n):
        yield x * x

squares_gen = get_squares(1000000)  # Does not allocate memory for all squares
for square in squares_gen:
    print(square)
```

* Reduce Boilerplate Code: It helps us to avoid writing explicit loops, making our code cleaner and more readable.

**Usage of map()**
* map()  is used in a function that takes one or more arguments and returns a value.

* We can pass multiple iterables to map(), and the function must accept as many arguments as there are iterables.

```python
def square(x):
    return x * x

numbers = [1, 2, 3, 4]
squared_numbers = map(square, numbers)
print(list(squared_numbers))

squared_numbers = map(lambda x: x * x, numbers)
print(list(squared_numbers))

list1 = [1, 2, 3]
list2 = [4, 5, 6]

added_numbers = map(lambda x, y: x + y, list1, list2)
print(list(added_numbers))
```

**10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

Ans: The map(), reduce(), and filter() functions in Python are all used for processing iterables, but they serve different purposes and work in distinct ways.

**map()**

Applies a function to every item in an iterable and returns a new iterable map object containing the results. When we want to transform data, such as applying a function to each element in a list.


```python
numbers = [1, 2, 3, 4]
squared = map(lambda x: x ** 2, numbers)  # Squares each number
print(list(squared))
```

**filter()**

Filters elements from an iterable based on a function that returns True or False, returning only those elements for which the function returns True.
It is uses, when we want to select a subset of data that meets certain criteria.

```python
numbers = [1, 2, 3, 4, 5]
even_numbers = filter(lambda x: x % 2 == 0, numbers)  # Filters even numbers
print(list(even_numbers))
```

**reduce()**

Reduces an iterable to a single cumulative value by applying a binary function means a function that takes two arguments cumulatively to the items of the iterable.
it is used, When we want to aggregate data, such as summing a list of numbers or multiplying them together.

```python
from functools import reduce

numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)  # Multiplies all numbers together
print(product)
```

**Major Difference**

map()
* Returns an iterable with transformed elements.
* Takes a function that operates on a single element.
* Use it to transform each element in an iterable.

filter()
* Returns an iterable with filtered elements based on a condition.
* Takes a function that returns True or False.
* Use it to select elements that meet certain criteria.

reduce()
* Returns a single cumulative value derived from the iterable.
* Takes a function that operates on two elements at a time.
* Use it to aggregate elements into a single value.
